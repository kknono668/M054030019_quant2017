# Q1. 寫一個函式，抓取 Google Finance 的股價歷史資料

```python
def data_finance(symbol,source,datetime):
    import io
    import pandas as pd
    import requests
    
    if source == 'google':
        url = "https://finance.google.com/finance/historical"
    elif source == 'yahoo':
        url = "https://finance.yahoo.com/quote/?q=/history"

    params = {}
    params['q'] = symbol
    params['startdate'] = datetime
    params['output'] = 'csv'

    r = requests.get(url,params = params)
    OHLC_data = pd.read_csv(io.StringIO(r.text))
    return OHLC_data
```

```python
df = data_finance('TSLA','google','2016-01-01')
print(df.head())

df = data_finance('AAPL','yahoo','2016-01-01')
print(df.head()) #本來抓得到ＱＱ 
```
